```python
# ANOVA Analysis
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.formula.api as smf
import numpy as np

anova_results = {}

# Loop through each emotion
for emotion in df['Emotion'].unique():
    # Subset data for the current emotion
    emotion_data = df[df['Emotion'] == emotion]
    
    # Fit the OLS model and calculate the ANOVA table
    model = smf.ols('Value ~ Performance', data=emotion_data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    
    # Store the result
    anova_results[emotion] = anova_table

# Print ANOVA tables for all emotions
for emotion, table in anova_results.items():
    print(f"ANOVA Table for {emotion}:\n{table}\n")
    print(" ")

# Performance = Between groups: k -1 (k = 2 groups) -> df = 1
# Residulas = Within groups: N - k
```

### Violation of normality of residuals, perform Kruskal-Wallis Test instead:

```python
# Kruskal-Wallis Test

from scipy.stats import kruskal
import pandas as pd

kruskal_results = {}

# Loop through each emotion
for emotion in df['Emotion'].unique():
    # Subset data for the current emotion
    emotion_data = df[df['Emotion'] == emotion]
    
    # Split data into groups (Human vs AI)
    human_values = emotion_data[emotion_data['Performance'] == 'Human']['Value']
    ai_values = emotion_data[emotion_data['Performance'] == 'AI']['Value']
    
    # Perform Kruskal-Wallis test
    stat, p = kruskal(human_values, ai_values)
    
    # Store the result
    kruskal_results[emotion] = {'H-Statistic': stat, 'p-value': p}

# Print Kruskal-Wallis results for all emotions
for emotion, result in kruskal_results.items():
    print(f"Kruskal-Wallis Test for {emotion}:")
    print(f"  Statistic = {result['H-Statistic']}, p-value = {result['p-value']}\n")
```

#### Calculate CI

```python
import numpy as np
from scipy import stats

# Function to calculate confidence intervals
def calculate_confidence_interval(data, confidence=0.95):
    """Calculate the confidence interval of a sample."""
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(len(data))
    margin_of_error = std_err * stats.t.ppf((1 + confidence) / 2., len(data) - 1)
    return mean - margin_of_error, mean + margin_of_error

# Function for bootstrapping
def bootstrap(data, num_iterations=1000):
    """Bootstrap resampling to calculate a confidence interval."""
    sample_means = []
    for _ in range(num_iterations):
        resampled_data = np.random.choice(data, size=len(data), replace=True)
        sample_means.append(np.mean(resampled_data))
    lower = np.percentile(sample_means, 2.5)
    upper = np.percentile(sample_means, 97.5)
    return lower, upper

kruskal_results = {}

# Loop through each emotion
for emotion in df['Emotion'].unique():
    # Subset data for the current emotion
    emotion_data = df[df['Emotion'] == emotion]
    
    # Split data into groups (Human vs AI)
    human_values = emotion_data[emotion_data['Performance'] == 'Human']['Value']
    ai_values = emotion_data[emotion_data['Performance'] == 'AI']['Value']
    
    # Perform Kruskal-Wallis test
    stat, p = kruskal(human_values, ai_values)
    
    # Calculate Confidence Intervals for Human and AI data
    human_ci = calculate_confidence_interval(human_values)
    ai_ci = calculate_confidence_interval(ai_values)
    
    # Calculate Bootstrapped Confidence Intervals for Human and AI data
    human_bootstrap = bootstrap(human_values)
    ai_bootstrap = bootstrap(ai_values)
    
    # Store the result
    kruskal_results[emotion] = {
        'H-Statistic': stat, 
        'p-value': p,
        'Human CI': human_ci, 
        'AI CI': ai_ci,
        'Human Bootstrap CI': human_bootstrap,
        'AI Bootstrap CI': ai_bootstrap
    }

# Print Kruskal-Wallis results, Confidence Intervals, and Bootstrap results
for emotion, result in kruskal_results.items():
    print(f"Kruskal-Wallis Test for {emotion}:")
    print(f"  Statistic = {result['H-Statistic']}, p-value = {result['p-value']}")
    print(f"  Human Confidence Interval: {result['Human CI']}")
    print(f"  AI Confidence Interval: {result['AI CI']}")
    print(f"  Human Bootstrap CI: {result['Human Bootstrap CI']}")
    print(f"  AI Bootstrap CI: {result['AI Bootstrap CI']}\n")
```
